```python
from sklearn.datasets import load_svmlight_file
import numpy as np
from pprint import pprint
import matplotlib.pyplot as plt
```

```python
Xte, Yte = load_svmlight_file('../aclImdb/test/labeledBow.feat')
Xtr, Ytr = load_svmlight_file('../aclImdb/train/labeledBow.feat')
```

```python
print(Xte.shape)
print(Xtr.shape)
```

```python
Ytr_int = np.array([int(y) for y in Ytr])
Yte_int = np.array([int(y) for y in Yte])
```

```python
unique_train, counts_train = np.unique(Ytr_int, return_counts=True)
unique_test, counts_test = np.unique(Yte_int, return_counts=True)
```

```python
colors = ['r', 'r', 'r', 'r', 'g', 'g', 'g', 'g']
plt.bar(unique_train, counts_train, color=colors)
plt.xlabel("Review Score")
plt.ylabel("Number of Reviews")
plt.savefig("train_review_dist.png", bbox_inches='tight')
plt.show()

colors = ['r', 'r', 'r', 'r', 'g', 'g', 'g', 'g']
plt.bar(unique_test, counts_test, color=colors)
plt.xlabel("Review Score")
plt.ylabel("Number of Reviews")
plt.savefig("test_review_dist.png", bbox_inches='tight')
plt.show()
```
